# Linear Predictive Coding

`Membaca data dengan frekuensi 16000Hz`

```python
def initialize(mp):
  signal,sr  = lbr.load(mp,duration=3,sr=16000)
  return signal,sr
```

**membangun fungsi LPF**

*dimana LPF akan menyaring sinyal frekuensi tinggi dan meneruskan sinyal frekuensi rendah yang diinginkan*

```python
def lowPassFilter(signal, coeff=0.97): #LPF
	return np.append(signal[0], signal[1:] - coeff * signal[:-1])
```

**PreEmphasis = *penyaringan ; mendapatkan bentuk spectral frekuensi sinyal yang lebih halus***

```python
def preEmphasis(mp):
	#inisialisasi sinyal awal
	signal , sr = initialize(mp)
 
	#Pre-emphasis dibangun diatas fungsi LPF
	preEmphasis = 0.97
	emphasizedSignal = lowPassFilter(signal)
	return emphasizedSignal, signal , sr
```

**OVERLAP = *memberikan hasil spektrum LPC yang akan berkorelasi tiap bagiannya***

```python
#sinyal yang telah di preEmphasis diblok menjadi beberapa bagian dengan jumlah sampel N

def framing(sr,signal):	
	#membagi sinyal ke dalam bentuk frame
	windowSize = 0.4 
	windowStep = 0.01 
	overlap = int(sr*windowStep)
	frameSize = int(sr*windowSize)
	numberOfframes = int(np.ceil(float(np.abs(len(signal) - frameSize)) / overlap ))

	#membentuk frame dengan array 2dimensi
	frames = np.ndarray((numberOfframes,frameSize))
	#memasukkan sampel kedalam frame
	for k in range(0,numberOfframes):
		for i in range(0,frameSize):
			if((k*overlap+i)<len(signal)):
				frames[k][i]=signal[k*overlap+i]
			else:
				frames[k][i]=0
	return frames,frameSize
```

```python
#MODEL WINDOW

windowSize = 0.4 
frameSize = int(sr*windowSize)
window = np.hamming(frameSize)

def hamming(frames,frameSize):
	#Windowing dengan Hamming
	#y[n] = s[n] * w[n] \\ (secara manual) 
	frames*=np.hamming(frameSize)
	return frames
```

`ANALISIS AUTOKORELASI`

```python
def autocorrelation(hammingFrames):
	KorelasiFrames=[]
	for k in range(len(hammingFrames)):
		KorelasiFrames.append(np.correlate(hammingFrames[k],hammingFrames[k],mode='full'))
		#correlateFrames adalah setiap frame setelah windowing dan autokorelasi 
	yolo =  KorelasiFrames[len(KorelasiFrames)/2:]
	return yolo
```

```python
#fungsi melihat pendekatan prediksi LPC
def autocor(signal, k):

    if k == 0:
        return np.sum(signal**2)
    else:
        return np.sum(signal[k:]*signal[:-k])
```

*membangun fungsi levinson-durbin menyelesaikan sistem persamaan linier orde-n*

```python
from scipy.linalg import solve_toeplitz, toeplitz

def levinsonDurbin(correlateFrames):
	#normalizedCF = preprocessing.normalize(correlateFrames, norm='l2')
	fit = levinson_durbin(correlateFrames,3)
	print(fit.numerator[1:])
 
def Levinson(w_sig,p):
    r_list = [autocor(w_sig,i) for i in range(p)]
    b_list = [autocor(w_sig,i) for i in range(1,p+1)]
    LPC = solve_toeplitz((r_list,r_list),b_list)
    return LPC
```

`ANALISIS LPC dengan koefisien cepstral`

```python
#Proses signal dan determines koefisien lpc(dengan fungsi autocorrelation)

def LPC_autocorrelation(order=16):
	folder = '/content/drive/MyDrive/aud/'
	amount = 39
	for x in range(1,int(amount)+1):
		mp = folder+str(x)+'.mp3'
		print('jenis rekaman:',mp)
		
		#preEmhasis
		emphasizedSignal,signal,sr = preEmphasis(mp)
		length = emphasizedSignal.size

		#autocorrelation , Fast Fourier Transform (FFT)
		autocorrelation = sig.fftconvolve(emphasizedSignal, emphasizedSignal[::-1])
		#koefisien autocorrelation 
		autocorr_coefficients = autocorrelation[autocorrelation.size//2:][:(order + 1)]
		
		#metode levinson_durbin dengan rumus toeplitz
		lpc_coefficients_levinson = levinson_durbin(autocorr_coefficients,3)
		print('koefisien korelasi dengan metode levinson:',lpc_coefficients_levinson.numerator)
		
		#Matriks Toeplitz
		R = linalg.toeplitz(autocorr_coefficients[:order])
		lpc_coefficients = np.dot(linalg.inv(R), autocorr_coefficients[1:order+1])

		#Ekstrak fitur LPC
		lpc_fitur=[]
		for i in lpc_coefficients:
			lpc_fitur.append(i)
		print('lpc fitur;',lpc_fitur)
```

# k-Nearest Neighbor (KNN)

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
```

*membuat data target menjadi bentuk label encoding*

```python
lb = LabelEncoder()
lb.fit(y_train)

y_train = lb.transform(y_train)
y_test = lb.transform(y_test)
```

*melakukan standarisasi pada variabel prediktor*

```python
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

*membangun model dengan membandingkan nilai k (1,15)*

`dimana 0 < k < 16`

```python
test_skor = []
train_skor = []

for i in range(1,16):

    knn = KNeighborsClassifier(i)
    knn.fit(X_train,y_train)
    
    train_skor.append(knn.score(X_train,y_train))
    test_skor.append(knn.score(X_test,y_test))
```

*identifikasi nilai maks variabel prediktor setiap percobaan nilai k dalam model*

```python
maks_train_skor = max(train_skor)
train_skor_ind = [i for i, v in enumerate(train_skor) if v == maks_train_skor]
print('Nilai Maks Skor Variabel Prediktor {} % dengan k = {}'.format(maks_train_skor*100,list(map(lambda x: x+1, train_skor_ind))))
```

*identifikasi nilai maks variabel target setiap percobaan nilai k dalam model*

```python
maks_test_skor = max(test_skor)
test_skor_ind = [i for i, v in enumerate(test_skor) if v == maks_test_skor]
print('Nilai Maks Skor Variabel Target {} % dengan k = {}'.format(maks_test_skor*100,list(map(lambda x: x+1, test_skor_ind))))
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

kfolds = KFold(n_splits=5, shuffle=True, random_state=123)
k_range = range(1, 20)

k_scores = []

for k in k_range:
    # run KNeighborsClassifier dengan k tetangga (1-19)
    knn = KNeighborsClassifier(n_neighbors=k)
    # menggunakan cross_val_score untuk KNeighborsClassifier dengan k tetangga (1-19)
    scores = cross_val_score(knn, X_train, y_train, cv=kfolds, scoring='accuracy')
    # memasukkan rata-rata scores untuk setiap k tetangga dalam list
    k_scores.append(scores.mean())

k_scores = [round(num, 2) for num in k_scores]
k = list(range(1,len(k_scores)))
df = pd.DataFrame(list(zip(k,k_scores)),columns =['K_Folds', 'K_Scores'])
df.sort_values(['K_Scores'],ascending=False).head()
```

```python
def macroavg_accuracy(y_test, y_pred):
    # axis = 0 (TP + TN)
    acc_x = np.sum(np.logical_and(y_test, y_pred), axis = 0)
    # axis = 0 (TP + TN + FP + FN)
    acc_y = np.sum(np.logical_or(y_test, y_pred), axis = 0)
    # (TP + TN) / (TP + TN + FP + FN)
    return np.mean(acc_x/acc_y)

acc_value = macroavg_accuracy(y_test, y_pred)
```
